```python
base_url = 'https://www.flipkart.com/search?q=shoes&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_2_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_2_0_na_na_na&as-pos=2&as-type=TRENDING&suggestionId=shoes&requestId=011fd47a-5c22-43dc-9629-e70dc05707b2&page='
```

```python
## Links
links = []
for i in range(2, 30):
    page = requests.get(base_url + str(i)).text
    soup = BeautifulSoup(page)
    link_ = soup.find_all('a',class_='_2mylT6')
    # print(link_)
    for i in link_:
        l = i.get('href')
        u = 'https://www.flipkart.com'
        url = u+l
        links.append(url)

print("Total No of Shoes: ", len(links))
```

```python
shoe_name
```

```python
name = []
color = []
gen = []
brand = []
price = []
offer = []
rating = []
no_of_ratings = []
no_of_reviews = []
nm=[]
count = 0
for i in links:
    data=requests.get(i)
    soup=BeautifulSoup(data.text)
    title = soup.findAll('span',class_='_35KyD6')
    
    title = title[0].text
    name.append(title)
    colour = re.findall(r'[(](.*)[)]',title)
    colour = colour[0]
    color.append(colour)
    name = re.findall(r'(.*)For',title)
    name1 = name[0]
    nm.append(name1)
    
    gender = re.findall(r'For\s*(\w+)',title)
    gender1 = gender[0]
    gen.append(gender1)
    
    brand_ = soup.findAll('span',class_='_2J4LW6')
    brand_ = brand_[0].text
    brand_ = re.findall(r'\w+',brand_)
    brand.append(brand_)
    
    price_ = soup.findAll('div',class_='_1vC4OE _3qQ9m1')
    price_ = re.findall(r'>\W(.*)<',str(price_))
    price.append(price_[0])
    
    offer_ = soup.findAll('div',class_='VGWI6T')
    offer_ = re.findall(r'(\d+%)',str(offer_))
    if len(offer_) == 1:
        offer.append(offer_[0])
    else:
        offer.append(np.nan)
        
    rating_ = soup.findAll('div',class_='hGSR34 bqXGTW')
    if rating_:
        rating.append(rating_[0])
    else:    
        rating.append(np.nan)
  
   
    no_of_ratings_ = soup.findAll('span',class_='_38sUEc')
    
    if no_of_ratings_:
        rating_review_group = re.search(r'([0-9,]+) ratings and ([0-9,]+)', no_of_ratings_[0].text)  
        no_of_ratings.append(rating_review_group.group(1))
        no_of_reviews.append(rating_review_group.group(2))
    else:
        no_of_ratings.append(np.nan)
        no_of_reviews.append(np.nan)

    count += 1
```

```python
shoe=pd.DataFrame()
shoe['Name']=nm
shoe['Colour']=color
shoe['Gender']=gen
shoe['Brand']=brand
shoe['Price']=price
shoe['Offer']=offer
shoe['Ratings']=rating
shoe['No of Ratings']=no_of_ratings
shoe['No of Reviews']=no_of_reviews
```
